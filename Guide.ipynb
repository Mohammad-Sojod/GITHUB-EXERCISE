# STEP-BY-STEP GUIDE OF THIS PRACTICE

### BEFORE
- create new folder and open in vscode
- create virtual environment: python -m venv nameofenv 
- Activate it: \nameofenv\Scripts\Activate.ps1

### DOWNLOAD A LIBRARY
- Write in terminal: pip install nameoflibrary (for example numpy)

## -> REPEATED PROCEDURE REGARDING GIT & GITHUB
- Of course create .gitignore file and paste the code from pyhton .gitignore in it.
- git init
- git add file1 file2 
- git commit -m "message here"
<br>
### -> Now we create connection with github
- git remote add origin https-link-here
- optional: git remote -v (now you can see remote connections created and procedures you can do)
<br>
### -> create a branch and push code to github repository
- git branch -M main (or manually create one)
- git push -u remote-connection-name branch-name
<br>
# NOTE: 
When creating a new branch, by default the new branch Copies current branch's latest snapshot (all files and commit history).<br>


## LET'S CREATE MORE BRANCHES FOR PRACTICES
- git branch name-of-new-branch
- add a file or edit the main just for testing.
- git add . (we added everything since .gitignore is available and handles what not to track and commit)
- git commit -m "message here"
- git push -u origin new-branch-name
<br>
### Now in vscode you will see that that the new branch was added

## LETS MERGE(PULL) DIFFERENT BRANCHES INTO THE MAIN BRANCH


- modify main.py file in the first branch - add - commit - push
- same but in the second branch
- now inside the github repository
